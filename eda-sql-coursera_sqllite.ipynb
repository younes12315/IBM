## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
query = '''
    SELECT *
    FROM SPACEXTBL
'''
data= pd.read_sql_query(query, con)
data
```

```python
query = '''
    SELECT DISTINCT Launch_Site
    FROM SPACEXTBL
'''
df_launchsite = pd.read_sql_query(query, con)
df_launchsite
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
query = '''
    SELECT  Launch_Site
    FROM SPACEXTBL
    WHERE Launch_Site LIKE 'CCA%'
    LIMIT 5
'''
df = pd.read_sql_query(query, con)
df
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
query = '''
    SELECT  sum(PAYLOAD_MASS__KG_) as "Total payload mass launched by NASA" 
    FROM SPACEXTBL
    WHERE Customer = "NASA (CRS)"
    
'''
df = pd.read_sql_query(query, con)
df
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
query = '''
    SELECT  AVG(PAYLOAD_MASS__KG_) as "Average payload Carried by booster version F9 v1.1" 
    FROM SPACEXTBL
    WHERE Booster_Version = "F9 v1.1"
    
'''
df = pd.read_sql_query(query, con)
df
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
query = '''
    SELECT  min(Date) as "the first succesful landing in ground pad was achieved " 
    FROM SPACEXTBL
    WHERE Landing_Outcome = "Success" AND Landing_Outcome = "Success (ground pad)"
    
'''
df = pd.read_sql_query(query, con)
df
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
query = '''
    SELECT  Booster_Version
    FROM SPACEXTBL
    WHERE 4000<PAYLOAD_MASS__KG_<6000 AND Landing_Outcome = "Success (drone ship)"
    
'''
df = pd.read_sql_query(query, con)
df
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python

query = '''
    SELECT
    SUM(CASE WHEN Mission_Outcome = 'Failure (in flight)' THEN 1 ELSE 0 END) as "Failure (in flight)",
    SUM(CASE WHEN Mission_Outcome IN ("Success (payload status unclear)","Success" ) THEN 1 ELSE 0 END) as "Success"
FROM SPACEXTBL
    
'''
df = pd.read_sql_query(query, con)
df
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
query = '''
SELECT 
     Booster_Version
FROM SPACEXTBL
WHERE PAYLOAD_MASS__KG_ IN (SELECT
    max(PAYLOAD_MASS__KG_)
FROM SPACEXTBL)
    
    
'''
df = pd.read_sql_query(query, con)
df
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python
query = '''
SELECT 
    SUBSTR(Date, 6, 2) AS "Month",
    Landing_Outcome AS "Failure landing_outcomes in drone ship",
    Booster_Version,
    Launch_Site AS "Launch_site for the month in 2015"
FROM SPACEXTBL
WHERE 
    Landing_Outcome LIKE '%(drone ship)' 
    AND
    (Landing_Outcome = 'Failure (drone ship)' OR Landing_Outcome = 'Precluded (drone ship)')
    AND
    SUBSTR(Date, 1, 4) = '2015';



    
'''
df = pd.read_sql_query(query, con)
df
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
query = '''
SELECT   
    Landing_Outcome,
    COUNT(Landing_Outcome) AS Outcome_Count
FROM SPACEXTBL
WHERE 
    Date >= '2010-06-04' AND Date < '2017-03-20'
    AND (
        (Landing_Outcome LIKE '%(drone ship)' OR Landing_Outcome LIKE '%(ground pad)')
        AND (
            Landing_Outcome = 'Failure (drone ship)' 
            OR Landing_Outcome = 'Precluded (drone ship)' 
            OR Landing_Outcome = 'Success (ground pad)'
        )
    )
GROUP BY Landing_Outcome
ORDER BY Date DESC;




    
'''
df = pd.read_sql_query(query, con)
df
```
